# 03 - Metrics Demo (PSD + Allan)

This notebook demonstrates using the package.


```python
import yaml, numpy as np
from qgrav.bench_ifo import generate_virtual_ifo
from qgrav.algorithms import estimate_displacement_improved
from qgrav.metrics import compute_psd, allan_deviation_overlapping
cfg = yaml.safe_load(open('../configs/example.yaml','r'))
b = cfg['bench_virtual_ifo']
data = generate_virtual_ifo(
    wavelength_m=b['wavelength_m'], sample_rate_hz=b['sample_rate_hz'], duration_s=b['duration_s'],
    displacement_sines=b['displacement_sines'], measurement_noise_std=b['measurement_noise_std'],
    amplitude=b.get('amplitude',1.0), dc_offset=b.get('dc_offset',0.0),
    offset_drift_std_per_s=b.get('offset_drift_std_per_s',0.0), amplitude_drift_std_per_s=b.get('amplitude_drift_std_per_s',0.0),
    seed=b.get('seed',0)
)
fs = float(b['sample_rate_hz'])
x_hat = estimate_displacement_improved(data['I_meas'], data['Q_meas'], wavelength_m=b['wavelength_m'])['x_hat']
psd = compute_psd(x_hat, fs)
taus = np.logspace(np.log10(1/fs), np.log10(0.25*b['duration_s']), 20)
adev = allan_deviation_overlapping(x_hat, fs, taus)
psd['f_hz'][:5], psd['psd'][:5], adev['taus_s'][:5], adev['adev'][:5]
```
