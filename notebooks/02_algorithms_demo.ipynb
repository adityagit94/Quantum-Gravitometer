# 02 - Algorithms Comparison Demo

This notebook demonstrates using the package.


```python
import numpy as np
import yaml

from qgrav.algorithms import estimate_displacement_baseline, estimate_displacement_improved
from qgrav.bench_ifo import generate_virtual_ifo

cfg = yaml.safe_load(open("../configs/example.yaml"))
b = cfg["bench_virtual_ifo"]
data = generate_virtual_ifo(
    wavelength_m=b["wavelength_m"],
    sample_rate_hz=b["sample_rate_hz"],
    duration_s=b["duration_s"],
    displacement_sines=b["displacement_sines"],
    measurement_noise_std=b["measurement_noise_std"],
    amplitude=b.get("amplitude", 1.0),
    dc_offset=b.get("dc_offset", 0.0),
    offset_drift_std_per_s=b.get("offset_drift_std_per_s", 0.0),
    amplitude_drift_std_per_s=b.get("amplitude_drift_std_per_s", 0.0),
    seed=b.get("seed", 0),
)
I, Q = data["I_meas"], data["Q_meas"]
x_true = data["x_true"]
lam = b["wavelength_m"]
x_b = estimate_displacement_baseline(I, Q, wavelength_m=lam)["x_hat"]
x_i = estimate_displacement_improved(I, Q, wavelength_m=lam)["x_hat"]
rmse_b = float(np.sqrt(np.mean((x_b - x_true) ** 2)))
rmse_i = float(np.sqrt(np.mean((x_i - x_true) ** 2)))
rmse_b, rmse_i, rmse_i / rmse_b
```
